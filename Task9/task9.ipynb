```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset (keep CSV in same folder or give full path)
df = pd.read_csv('testset.csv') 

# Step 3: Data Preprocessing
df['datetime_utc'] = pd.to_datetime(df['datetime_utc'])
df = df.sort_values('datetime_utc')
df.set_index('datetime_utc', inplace=True)

# Check column names
print("Columns in dataset:", df.columns)

# Fix column names (remove spaces if any)
df.columns = df.columns.str.strip()

# Fix column names (remove leading underscore if any)
df.columns = df.columns.str.lstrip('_')

# Step 4: Line Graph (Temperature)
plt.figure(figsize=(12,5))
plt.plot(df['tempm'], label='Temperature')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()

# Step 5: Trend Line (Rolling Average)
plt.figure(figsize=(12,5))
plt.plot(df['tempm'], label='Original')
plt.plot(df['tempm'].rolling(window=30).mean(), label='Trend (30-day avg)')
plt.title('Temperature Trend Line')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()

# Step 6: Area Chart (Humidity)
plt.figure(figsize=(12,5))
plt.fill_between(df.index, df['hum'], alpha=0.5)
plt.title('Humidity Area Chart')
plt.xlabel('Date')
plt.ylabel('Humidity')
plt.show()

# Step 7: Multiple Feature Visualization
plt.figure(figsize=(12,5))
plt.plot(df['tempm'], label='Temperature')
plt.plot(df['hum'], label='Humidity')
plt.plot(df['pressurem'], label='Pressure')
plt.title('Weather Parameters Over Time')
plt.legend()
plt.show()
```
