```python
# Wordtree Visualization

import matplotlib.pyplot as plt

# Text document
text = """
Python is a powerful programming language
Python is widely used in data science
Python helps in machine learning and visualization
Python is easy to learn
Python is very popular in the world
"""

root_word = "Python"

# Split sentences
sentences = text.split("\n")

branches = []

for s in sentences:
    words = s.split()
    if root_word in words:
        index = words.index(root_word)
        branch = " ".join(words[index+1:])
        branches.append(branch)

plt.figure(figsize=(6,4))

# Draw root
plt.text(0.1, 0.5, root_word, fontsize=20, fontweight='bold')

y = 0.8

for b in branches:
    plt.plot([0.2,0.6],[0.5,y])
    plt.text(0.6,y,b)
    y -= 0.12

plt.title("WordTree Visualization")
plt.axis("off")

plt.show()
```

```python
# TextArc Visualization
import matplotlib.pyplot as plt
import numpy as np

text = """
Data visualization helps people understand complex data visualization techniques
include graphs charts dashboards and storytelling visualization helps decision making
"""

words = text.split()

n = len(words)

# Angle positions
angles = np.linspace(0, 2*np.pi, n, endpoint=False)

radius = 5

x = radius * np.cos(angles)
y = radius * np.sin(angles)

plt.figure(figsize=(8,8))

# Draw circle boundary
circle = plt.Circle((0,0), radius, fill=False, linestyle='dashed')
plt.gca().add_patch(circle)

# Plot words around circle
for i, word in enumerate(words):
    plt.text(x[i], y[i], word, ha='center', va='center')

# Draw arcs for repeated words
for i in range(n):
    for j in range(i+1, n):
        if words[i] == words[j]:
            plt.plot([x[i], x[j]], [y[i], y[j]], alpha=0.6)

plt.title("TextArc Circular Visualization")

plt.axis('equal')
plt.axis('on')

plt.show()
```

```python
# Arc Diagram Visualization
import matplotlib.pyplot as plt
import numpy as np

# Node labels
nodes = ["Part A1", "Part A2", "Transition", "Part B1", "Part B2"]

# Connections (start_index, end_index)
edges = [
    (0,1),
    (1,2),
    (2,3),
    (3,4),
    (0,3)
]

x = np.arange(len(nodes))

plt.figure(figsize=(8,5))

# draw nodes
colors = ["purple","steelblue","teal","limegreen","gold"]

for i,node in enumerate(nodes):
    plt.scatter(x[i],0,s=200,color=colors[i])
    plt.text(x[i],-0.1,node,ha='center')

# draw arcs
for i,(start,end) in enumerate(edges):

    mid = (x[start]+x[end])/2
    radius = abs(x[end]-x[start])/2

    theta = np.linspace(0,np.pi,100)

    x_arc = mid + radius*np.cos(theta)
    y_arc = radius*np.sin(theta)

    plt.plot(x_arc,y_arc,color=colors[start],linewidth=2)

plt.title("Arc Diagram ")

plt.axis("off")
plt.show()
```
