```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="whitegrid")
```

```python
import os
print(os.getcwd())
print(os.listdir())
```

```python
#Load Dataset
df = pd.read_csv("student-mat.csv")
df.head()
# print(df.info())
```

```python
df.tail()
```

```python
#Identify Categorical & Continuous Columns
print("Categorical Columns:")
print(df.select_dtypes(include=['object']).columns)

print("Continuous Columns:")
print(df.select_dtypes(include=['int64','float64']).columns)
```

Categorical vs. Categorical

```python
#1.STACKED BAR CHART
ct = pd.crosstab(df['school'], df['address'])

ct.plot(kind='bar', stacked=True)
plt.title("Stacked Bar Chart: School vs Address")
plt.xlabel("School")
plt.ylabel("Count")
plt.show()
```

```python
#2.GROUPED BAR CHART
ct.plot(kind='bar')
plt.title("Grouped Bar Chart: Higher vs Gender")
plt.show()
```

```python
#3️.SEGMENTED BAR CHART (100% Stacked)
ct_percent = ct.div(ct.sum(axis=1), axis=0)

ct_percent.plot(kind='bar', stacked=True)
plt.title("Segmented Bar Chart (Percentage)")
plt.ylabel("Percentage")
plt.show()
```

```python
#4.MOSAIC PLOT
from statsmodels.graphics.mosaicplot import mosaic
mosaic(df, ['internet', 'paid'])
plt.title("Mosaic Plot: Internet vs Paid")
plt.show()
```

Continuous vs. Continuous

```python
#5.SCATTERPLOT WITH FIT LINE
sns.lmplot(x='G1', y='G3', data=df)
plt.title("Scatterplot with Fit Line (G1 vs G3)")
plt.show()
```

Categorical vs. Continuous

```python
#6.BAR CHART (Mean Summary)
df.groupby('reason')['G3'].mean().plot(kind='bar')
plt.title("Average Final Grade by Reason for Choosing School")
plt.ylabel("Mean G3")
plt.show()
```

```python
#7.GROUPED KDE PLOT
sns.kdeplot(data=df, x='age', hue='sex', fill=True)
plt.title("Grouped Kernel Density Plot")
plt.show()
```

```python
#8.BOX PLOT
sns.boxplot(x='internet', y='G1', data=df)
plt.title("Box Plot (G3 by Gender)")
plt.show()
```

```python
#9.VIOLIN PLOT
sns.violinplot(x='address', y='goout', data=df)
plt.title("Violin Plot (Travel Time by Address)")
plt.show()
```

```python
#10.RIDGELINE PLOT
import joypy
joypy.joyplot(df, by='sex', column='age')
plt.title("Ridgeline Plot (G3 by Gender)")
plt.show()
```

```python
#11.BEESWARM PLOT
sns.swarmplot(x='sex', y='G3', data=df)
plt.title("Beeswarm Plot (G3 by Gender)")
plt.show()
```
