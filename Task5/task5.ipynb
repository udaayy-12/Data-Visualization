```python
import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("cinema_tickets.csv")

# Remove rows where important columns are missing
df = df.dropna(subset=["City", "TheatreName", "MovieName", "TicketPrice","SeatType"])

# Create Treemap
fig = px.treemap(
    df,
    path=["City", "TheatreName", "MovieName","SeatType"],  # Hierarchy
    values="TicketPrice",                      # Size of boxes
    color="TicketPrice",                       # Color scale
    color_continuous_scale="Viridis",
    title="Cinema Ticket Sales TreeMap Display (4 Layers)"
)

# Improve layout
fig.update_layout(
    margin=dict(t=50, l=25, r=25, b=25)
)

fig.show()
```

Task 5b

```python
import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("cinema_tickets.csv")

# Replace NULL values so they appear in chart
df = df.fillna("Unknown")
df["TicketPrice"] = pd.to_numeric(df["TicketPrice"], errors="coerce").fillna(0)

# Group data with 4 hierarchy levels
df_grouped = df.groupby(
    ["City", "TheatreName", "MovieName", "SeatType"]
)["TicketPrice"].sum().reset_index()

# Create Sunburst with 4 layers
fig = px.sunburst(
    df_grouped,
    path=[
        "City",
        "TheatreName",
        "MovieName",
        "SeatType"
    ],
    values="TicketPrice",
    color="TicketPrice",
    color_continuous_scale="Viridis",
    title="Cinema Ticket Revenue Sunburst (4 Layers)"
)

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

fig.show()
```
