```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")
```

```python
#Load Your Dataset
df = pd.read_csv("best_selling_books.csv", encoding='latin1')
df.head()
# print(df.info())

```

```python
df.tail()
```

```python
#Identify Data Types
print("Categorical Columns:")
print(df.select_dtypes(include=['object']).columns)

print("Continuous Columns:")
print(df.select_dtypes(include=['int64','float64']).columns)
```

Task-2a

```python
#BAR CHART
plt.figure(figsize=(10,5))
df['Rating'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Bar Chart of Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#2.PIE CHART
plt.figure(figsize=(6,6))
df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart of Genre")
plt.ylabel("")
plt.show()
```

```python
#3.SCATTER PLOT
plt.figure(figsize=(8,5))
plt.scatter(df.index, df['reviews count'])
plt.title("Scatter Plot of Reviews Count")
plt.xlabel("id")
plt.ylabel("Reviews Count")
plt.show()
```

```python
#4.LINE PLOT
plt.figure(figsize=(8,5))
plt.plot(df['Print Length'])
plt.title("Line Plot of Print Length")
plt.xlabel("Reading age")
plt.ylabel("Print Length")
plt.show()

```

```python
#5.STRIP PLOT
plt.figure(figsize=(8,5))
sns.stripplot(x=df['reviews count'])
plt.title("Strip Plot of Reviews Count")
plt.show()
```

```python
#6.SWARM PLOT
plt.figure(figsize=(8,5))
sns.swarmplot(x=df['Print Length'])
plt.title("Swarm Plot of Print Length")
plt.show()
```

```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
#Load Employee Dataset
emp = pd.read_csv("employee.csv", encoding='latin1')
emp.head()
# print(emp.info())
```

```python
emp.tail()
```

```python
#Identify Categorical & Continuous Columns
print("Categorical Columns:")
print(emp.select_dtypes(include=['object']).columns)

print("Continuous Columns:")
print(emp.select_dtypes(include=['int64','float64']).columns)

```

```python
#1.BARCHART (Categorical â Education)
plt.figure(figsize=(8,5))
emp['Education'].value_counts().plot(kind='bar')
plt.title("Bar Chart of Education")
plt.xlabel("Education")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#2.PIE CHART (Categorical â Gender)
plt.figure(figsize=(6,6))
emp['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart of Gender")
plt.ylabel("")
plt.show()
```

```python
#3.HISTOGRAM (Continuous â Age)
plt.figure(figsize=(8,5))
plt.hist(emp['Age'], bins=20)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#4.DENSITY PLOT (Continuous â Age)
plt.figure(figsize=(8,5))
sns.kdeplot(emp['Age'], fill=True)
plt.title("Density Plot of Age")
plt.show()
```

```python
#5.RUG PLOT (Continuous â Age)
plt.figure(figsize=(8,5))
sns.rugplot(emp['Age'])
plt.title("Rug Plot of Age")
plt.show()
```
